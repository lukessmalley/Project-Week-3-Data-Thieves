# Given Spotify’s annual top 100 song charts, is there a marked preference for songs with specific features/metrics among Spotify users?

```python
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# import csv
csvfile = "datasets/top2018.csv"
df = pd.read_csv(csvfile)
df
```

```python
# select revelant columns
# dropping key and time_signature because of low variance
# and acousticness and liveness because they are not pertinent to the question
cols_to_drop = ["id", "key", "time_signature", "acousticness", "liveness"]
df_light = df.drop(cols_to_drop, axis=1)
df_light
```

```python
# calculate averages of each metric

# TOP 50
stats = df_light.describe().transpose()
stats
```

```python
fig, axs = plt.subplots(2, 3, figsize=(20, 10))
axs[0, 0].hist(df_light['danceability'], 30, range=[-0.1, 1.1], facecolor='gray', align='mid')
axs[0, 0].set_title('Danceability')
axs[0, 1].hist(df_light['loudness'], 30, range=[-60.5, 0.5], facecolor='gray', align='mid')
axs[0, 1].set_title('Loudness')
axs[0, 2].hist(df_light['energy'], 30, range=[-0.1, 1.1], facecolor='gray', align='mid')
axs[0, 2].set_title('Energy')
axs[1, 0].hist(df_light['speechiness'], 30, range=[-0.1, 1.1], facecolor='gray', align='mid')
axs[1, 0].set_title('Speechiness')
axs[1, 1].hist(df_light['instrumentalness'], 30, range=[-0.1, 1.1], facecolor='gray', align='mid')
axs[1, 1].set_title('Instrumentalness')
axs[1, 2].hist(df_light['tempo'], 30, range=[0, 220], facecolor='gray', align='mid')
axs[1, 2].set_title('Tempo')
plt.show()
```

## Conclusions:

- **Danceability:**
Scale 0-1: mean=0.71 with std=0.13.
More danceable is preferred.

- **Loudness:**
Scale -60-0: mean=-6 with std=1.7.
Louder is preferred.

- **Energy:**
Scale 0-1: mean=0.65 with std=0.14.
Neither excessive nor lacking energy is preferred.

- **Speechiness:**
Scale 0-1 (songs usually within the 0-0.66 range): mean=0.11 with std=0.1.
Less speechiness is preferred (p.e.: rap would be 0.33-0.66).

- **Instrumentalness:**
Scale 0-1 (instrumental only is usually >0.5): mean=0.001 with std=0.01.
Songs with vocals are definitely preferred.

- **Tempo:**
Usually 20-200: mean=119.
Songs in the _allegro_ region of tempo are preferred (not too fast but not slow).

- **Valence:**
Scale 0-1: mean=0.48 with std=0.2.
Both positive- and negative-sounding songs are equally preferred.

- **Mode:**
Scale 0-1: mean 0.59 with std=0.49.
Both major and minor modes are equally preferred, confirming valence data.


## Exceptions to the rule
What sets them apart?

```python
# calculate relative distances to mean for each metric
df_light["dist_danceability"] = abs((df_light["danceability"] - stats["mean"]["danceability"])/stats["mean"]["danceability"])
df_light["dist_energy"] = abs((df_light["energy"] - stats["mean"]["energy"])/stats["mean"]["energy"])
df_light["dist_loudness"] = abs((df_light["loudness"] - stats["mean"]["loudness"])/stats["mean"]["loudness"])
df_light["dist_tempo"] = abs((df_light["tempo"] - stats["mean"]["tempo"])/stats["mean"]["tempo"])
df_light["dist_valence"] = abs((df_light["valence"] - stats["mean"]["valence"])/stats["mean"]["valence"])
df_light["dist_instrumentalness"] = abs((df_light["instrumentalness"] - stats["mean"]["instrumentalness"])/stats["mean"]["instrumentalness"])
df_light["dist_speechiness"] = abs((df_light["speechiness"] - stats["mean"]["speechiness"])/stats["mean"]["speechiness"])

# sum distances
df_light["diff"] = df_light["dist_danceability"] + df_light["dist_energy"] + df_light["dist_loudness"] + df_light["dist_tempo"] + df_light["valence"] + df_light["instrumentalness"] + df_light["speechiness"]
df_light.sort_values(by=["diff"], ascending=False, inplace=True)

# get top 6 furthest from the mean
df_top6_furthest = df_light.head(6)
```

```python
# which 5 artists have a greater presence in the top chart?
df_top5_popular = df_light.groupby("artists", as_index=False).agg({"name":"count"})
df_top5_popular = df_top5_popular.sort_values(by=["name"], ascending=False)
df_top5_popular.columns = ["Artist", "Song count"]
df_top5_popular[df_top5_popular["Artist"]=="Billie Eilish"]
```

```python
df_top6_furthest = df_top6_furthest[["name", "artists", "danceability", "energy", "loudness", "speechiness", "instrumentalness", "valence", "tempo"]].copy()

df_top6_furthest["Song count"] = [1, 6, 1, 1, 1, 6]
dd = df_top6_furthest[["artists", "Song count"]]
dd.columns = ["Artist", "Song count"]
dd.sort_values(by=["Artist"], inplace=True)
df_top6_furthest.sort_values(by=["artists"], inplace=True)
```

```python
x = df_top6_furthest['artists'].sort_values()
x[28] = " " + x[28]

fig, axs = plt.subplots(2, 3, figsize=(20, 10))
axs[0, 0].plot(x, df_top6_furthest['danceability'], 'go--', linewidth=2, markersize=12, label='values')
axs[0, 0].plot(x, [stats["mean"]["danceability"]]*6, label='mean')
axs[0, 0].set_ylim([0, 1])
axs[0, 0].set_xticklabels(x, rotation=45)
axs[0, 0].set_title('Danceability', fontdict={'fontsize': 12})
axs[0, 1].plot(x, df_top6_furthest['loudness'], 'go--', linewidth=2, markersize=12, label='values')
axs[0, 1].plot(x, [stats["mean"]["loudness"]]*6, label='mean')
axs[0, 1].set_ylim([-60, 0])
axs[0, 1].set_xticklabels(x, rotation=45)
axs[0, 1].set_title('Loudness')
axs[0, 2].plot(x, df_top6_furthest['energy'], 'go--', linewidth=2, markersize=12, label='values')
axs[0, 2].plot(x, [stats["mean"]["energy"]]*6, label='mean')
axs[0, 2].set_ylim([0, 1])
axs[0, 2].set_xticklabels(x, rotation=45)
axs[0, 2].set_title('Energy')
axs[1, 0].plot(x, df_top6_furthest['speechiness'], 'go--', linewidth=2, markersize=12, label='values')
axs[1, 0].plot(x, [stats["mean"]["speechiness"]]*6, label='mean')
axs[1, 0].set_ylim([0, 1])
axs[1, 0].set_xticklabels(x, rotation=45)
axs[1, 0].set_title('Speechiness')
axs[1, 1].plot(x, df_top6_furthest['instrumentalness'], 'go--', linewidth=2, markersize=12, label='values')
axs[1, 1].plot(x, [stats["mean"]["instrumentalness"]]*6, label='mean')
axs[1, 1].set_ylim([0, 0.2])
axs[1, 1].set_xticklabels(x, rotation=45)
axs[1, 1].set_title('Instrumentalness')
axs[1, 2].plot(x, df_top6_furthest['tempo'], 'go--', linewidth=2, markersize=12, label='values')
axs[1, 2].plot(x, [stats["mean"]["tempo"]]*6, label='mean')
axs[1, 2].set_ylim([0, 220])
axs[1, 2].set_xticklabels(x, rotation=45)
axs[1, 2].set_title('Tempo')
fig.subplots_adjust(hspace=0.5)

plt.show()
```
